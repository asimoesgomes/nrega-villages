```python
def list_to_file(lista,file):
    final_string = ''
    for i, lis in enumerate(lista):
        if i==len(lista)-1:
            final_string += str(i)+','+str(lis)
        else:
            final_string += str(i)+','+str(lis)+'\n'
    f = open('/home/users/asimoes/data/intermediate/'+file,'w')
    f.write(final_string)
    f.close()
```

```python
def mixlist_to_file(lista1,lista2,file):
    final_string = ''
    i = 0
    for item1 in lista1:
        for item2 in lista2:
            final_string += str(i)+','+str(item1)+','+str(item2)+'\n'
            i += 1
    f = open('/home/users/asimoes/data/intermediate/'+file,'w')
    f.write(final_string)
    f.close()
```

#### Let's match by steps: 
####    STATE - State.Name/STATE_UT
####    DISTRICT - District.Name/DISTRICT
####    SUB-DISTRICT - Block.Name/SUB_DIST
####    VILLAGE - Panchayat.Name/NAME

```python
def items_to_file(typ,nrega_name,vil_name,nrega_df,vil_df):
    nrega  = nrega_df[~nrega_df[nrega_name].isnull()][nrega_name].sort_values().unique().tolist()
    list_to_file(nrega,typ+'_nrega.txt')
    vil  = vil_df[~vil_df[vil_name].isnull()][vil_name].sort_values().unique().tolist()
    list_to_file(vil,typ+'_vil.txt')
    mixlist_to_file(nrega,vil,typ+'.txt')
    return
```

```python
def dist_to_df(typ):
    dist = pd.read_csv('/home/users/asimoes/data/intermediate/'+typ+'_dist.txt')
    dist.columns = ['row_number','nrega','vil','dist']
    dist = dist[~dist.vil.isnull()]
    dist.drop('row_number',axis=1,inplace=True)
    dist.sort_values(['nrega','dist'],inplace=True)
    dist.drop_duplicates('nrega',inplace=True)
    return dist
```

## Merging Higher Levels Before Village

```python
nrega = nrega.merge(state_dist[['nrega','vil','dist']].rename(columns={'nrega':'State.Name','vil':'state_vil','dist':'state_dist'}),
               on='State.Name',how='left')
nrega = nrega.merge(district_dist[['nrega','vil','dist']].rename(columns={'nrega':'District.Name','vil':'district_vil','dist':'district_dist'}),
               on='District.Name',how='left')
nrega = nrega.merge(sub_district_dist[['nrega','vil','dist']].rename(columns={'nrega':'Block.Name','vil':'sub_dis_vil','dist':'sub_dis_dist'}),
               on='Block.Name',how='left')
```

```python
nrega['total_dist'] = nrega['state_dist']+nrega['district_dist']+nrega['sub_dis_dist']
```

```python
nrega['total_rel_dist'] = 0
for unit, dist in [('State.Name','state_dist'),('District.Name','district_dist'),('Block.Name','sub_dis_dist')]:
    name = dist.replace('_dist','')
    nrega[name+'_rel_dist'] = nrega[dist]/nrega[unit].str.len()
    nrega['total_rel_dist'] += nrega[name+'_rel_dist']
```

```python
nrega = nrega.sort_values('total_rel_dist',ascending=False)
nrega[['State.Name','state_vil','District.Name','district_vil','Block.Name','sub_dis_vil','total_rel_dist','total_dist']].drop_duplicates().reset_index(drop=True).iloc[400:,:].head(60)
# After 0.239 we already miss some correct matches for sure - let's start from this because we can filter after merging villages
```
